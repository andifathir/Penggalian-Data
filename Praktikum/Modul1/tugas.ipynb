```python
import random

def manipulasi_list(data):
    print("Data Awal:", data)

    # 1. Tambahkan 10 elemen numerik (int dan float) secara acak.
    for _ in range(10):
        elemen_acak = random.choice([random.randint(1, 100), random.uniform(1, 100)])
        data.append(elemen_acak)
    print("Setelah Menambahkan 10 Elemen Acak:", data)
    
    # 2. Hapus elemen di indeks sesuai NIM (ganjil/genap). 
    # (NIM genap, jadi kita akan hapus elemen dengan indeks genap)
    data = [elemen for idx, elemen in enumerate(data) if idx % 2 != 0]  # hanya indeks ganjil
    print("Setelah Menghapus Elemen dengan Indeks Genap:", data)
    
    # 3. Balik urutan elemen dalam list.
    data.reverse()
    print("Setelah Membalik Urutan Elemen:", data)
    
    # 4. Hapus elemen dengan nilai terbesar dari list tersebut.
    if data:  # Pastikan list tidak kosong
        data.remove(max(data))
    print("Setelah Menghapus Elemen Terbesar:", data)
    
    # 5. Urutkan elemen pada list secara descending.
    data.sort(reverse=True)
    print("Setelah Mengurutkan Secara Descending:", data)
    
    return data

# Contoh pemanggilan fungsi
data_awal = [1, 2, 3, 4, 5]
hasil = manipulasi_list(data_awal)
```

```python
# Membuat tuple yang berisi dictionary dengan 10 pasangan key-value mata kuliah dan nilai
mata_kuliah = (
    {"Matematika": 85},
    {"Fisika": None},  # Nilai kosong
    {"Kimia": 90},
    {"Biologi": 78},
    {"Pendidikan Agama": None},  # Nilai kosong
    {"Bahasa Indonesia": 88},
    {"Ekonomi": None},  # Nilai kosong
    {"Sejarah": 92},
    {"Geografi": 81},
    {"Seni Budaya": 76}
)

# Menggabungkan semua dictionary dalam tuple menjadi satu dictionary
data = {}
for item in mata_kuliah:
    data.update(item)

print("Data Awal:", data)

# 1. Hapus entri yang tidak lengkap atau memiliki nilai kosong
data = {key: value for key, value in data.items() if value is not None}
print("\nSetelah Menghapus Entri yang Memiliki Nilai Kosong:", data)

# 2. Hitung jumlah entri yang tersisa setelah penghapusan
jumlah_entri = len(data)
print("\nJumlah Entri yang Tersisa:", jumlah_entri)

# 3. Urutkan pasangan key-value berdasarkan nilai secara ascending
data_terurut = dict(sorted(data.items(), key=lambda item: item[1]))
print("\nSetelah Mengurutkan Berdasarkan Nilai Secara Ascending:", data_terurut)

# 4. Carilah mata kuliah dengan nilai tertinggi
mata_kuliah_tertinggi = max(data, key=data.get)
nilai_tertinggi = data[mata_kuliah_tertinggi]
print("\nMata Kuliah dengan Nilai Tertinggi:", mata_kuliah_tertinggi, "Nilai:", nilai_tertinggi)

# 5. Hitunglah rata-rata semua nilai mata kuliah
rata_rata = sum(data.values()) / len(data)
print("\nRata-rata Nilai Mata Kuliah:", rata_rata)
```

```python
import pandas as pd
import numpy as np

# Membuat DataFrame berdasarkan informasi yang diberikan
data = {
    'Brand': ['Avoskin', 'G2G', 'Skintific', 'Wardah', 'Emina', 'Somethinc', 'Azarine', 'Lacoco', 'Y.O.U', 'Madame Gie'],
    'Cabang': ['Surabaya', 'Jakarta Pusat', 'Bandung', 'Bandung', 'Jakarta', 'Bandung', 'Surabaya', 'Bandung', 'Jakarta Pusat', 'Surabaya'],
    'Harga (ribu)': [150, 75, 120, 85, 75, 120, 50, 90, 55, 50],
    'Stock': [175, 160, 120, 300, 230, 200, 180, 100, 180, 200],
    'Status': ['Sold', 'Available', 'Available', 'Sold', 'Available', 'Available', 'Available', 'Sold', 'Available', 'Sold']
}

df = pd.DataFrame(data)

# 1. Tampilkan ringkasan statistik deskriptif
print("Ringkasan Statistik Deskriptif:")
print(df.describe())

# 2. Tentukan 3 brand dengan penghasilan terbesar (Harga x Stock)
df['Penghasilan'] = df['Harga (ribu)'] * df['Stock']
top_3_brands = df.nlargest(3, 'Penghasilan')
print("\n3 Brand dengan Penghasilan Terbesar:")
print(top_3_brands[['Brand', 'Penghasilan']])

# 3. Kelompokkan brand dengan harga yang sama
grouped_by_price = df.groupby('Harga (ribu)')['Brand'].apply(list)
print("\nKelompok Brand dengan Harga yang Sama:")
print(grouped_by_price)

# 4. Tambahkan kolom kategori harga
Q1 = df['Harga (ribu)'].quantile(0.25)
Q3 = df['Harga (ribu)'].quantile(0.75)

def kategori_harga(row):
    if row['Harga (ribu)'] < Q1:
        return 'Murah'
    elif row['Harga (ribu)'] > Q3:
        return 'Mahal'
    else:
        return 'Sedang'

df['Kategori Harga'] = df.apply(kategori_harga, axis=1)

print("\nDataFrame dengan Kolom Kategori Harga:")
print(df[['Brand', 'Harga (ribu)', 'Kategori Harga']])
```
